# Imputasi Data

```python
miss_var = ['pisat', 'jenkel', 'diagfktp', 'jenispulang', 'kdsa', 'kdsr', 'kdsp', 'kdsi', 'kdsd', 'politujuan']
for feature in miss_var:
    df[feature].fillna(df[feature].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
median_0 = df[df['label'] == 0]['biaya'].median()
median_1 = df[df['label'] == 1]['biaya'].median()
median_0, median_1
```

```python
null_0 = df[df['label'] == 0]
null_0['biaya'].fillna(null_0['biaya'].median(), inplace=True)
null_0.isna().sum()
```

```python
null_1 = df[df['label'] == 1]
null_1['biaya'].fillna(null_1['biaya'].median(), inplace=True)
null_1
```

# Reducing categories of diagfktp

```python
df1 = df.copy()
df1['diagfktp'].unique()[:100]
```

```python
df1['char_diagfktp'] = df1['diagfktp'].str[0:2]
df1['char_diagfktp'] = df1['char_diagfktp'].str.upper()
sorted(df1['char_diagfktp'].unique())

for i in df1['char_diagfktp'].unique():
    if i.startswith('W'):
        print(i)
```

```python
key = sorted(df1['diagfktp'].str[0].str.upper().unique())
mapping = {}
for key in key:
    mapping[key] = None
mapping
```

```python
mapping['A'] = 1
mapping['B'] = 1
mapping['C'] = 2
mapping['E'] = 4
mapping['F'] = 5
mapping['G'] = 6
mapping['I'] = 9
mapping['J'] = 10
mapping['K'] = 11
mapping['L'] = 12
mapping['M'] = 13
mapping['N'] = 14
mapping['O'] = 15
mapping['P'] = 16
mapping['Q'] = 17
mapping['R'] = 18
mapping['S'] = 19
mapping['T'] = 19
mapping['W'] = 0
mapping['Z'] = 22
mapping['D0'] = 2
mapping['D1'] = 2
mapping['D2'] = 2
mapping['D3'] = 2
mapping['D4'] = 2
mapping['D5'] = 3
mapping['D6'] = 3
mapping['D7'] = 3
mapping['D8'] = 3
mapping['H0'] = 7
mapping['H1'] = 7
mapping['H2'] = 7
mapping['H3'] = 7
mapping['H4'] = 7
mapping['H5'] = 7
mapping['H6'] = 8
mapping['H7'] = 8
mapping['H8'] = 8
mapping['H9'] = 8
```

```python
mapping.pop('D')
mapping.pop('H')
```

```python
df1['char_diagfktp']
```

```python
diag = df1['char_diagfktp']
for idx, i in enumerate(diag):
    if i.startswith('D') or i.startswith('H'):
        pass
    else:
        diag.iloc[idx] = diag.iloc[idx][0]
```

```python
diag.unique()
```

```python
df1['diagfktp_encoded'] = diag.map(mapping)
```

# kdsa, kdsp, kdsr, kdsi, kdsd

```python
df3 = df.copy()
df3[['kdsa', 'kdsp', 'kdsr', 'kdsi', 'kdsd']]
```

```python
cols = ['kdsa', 'kdsp', 'kdsr', 'kdsi', 'kdsd']

for col in cols:
    print(df3[col].unique())
```

```python
for col in cols:
    df3[col + '_sliced'] = df3[col].str[0]
    print(col, df3[col + '_sliced'].unique())
```

```python
map_kdsa = {'N' : 0, 'S' : 1}
map_kdsp = {'N' : 0, 'Y' : 1}
map_kdsr = {'N' : 0, 'R' : 1}
map_kdsi = {'N' : 0, 'I' : 1}
map_kdsd = {'N' : 0, 'D' : 1}
```

```python
df3['kdsa_encoded'] = df3['kdsa_sliced'].map(map_kdsa)
df3['kdsp_encoded'] = df3['kdsp_sliced'].map(map_kdsp)
df3['kdsr_encoded'] = df3['kdsr_sliced'].map(map_kdsr)
df3['kdsi_encoded'] = df3['kdsi_sliced'].map(map_kdsi)
df3['kdsd_encoded'] = df3['kdsd_sliced'].map(map_kdsd)
```

```python
df['kdsa_encoded'] = df3['kdsa_sliced'].map(map_kdsa)
df['kdsp_encoded'] = df3['kdsp_sliced'].map(map_kdsp)
df['kdsr_encoded'] = df3['kdsr_sliced'].map(map_kdsr)
df['kdsi_encoded'] = df3['kdsi_sliced'].map(map_kdsi)
df['kdsd_encoded'] = df3['kdsd_sliced'].map(map_kdsd)
```
